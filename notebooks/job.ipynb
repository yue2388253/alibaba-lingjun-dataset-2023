```python
df.loc[:, 'gmt_job_submitted'] = pd.to_datetime(df['gmt_job_submitted'], format='%Y/%m/%d %H:%M')
df.loc[:, 'gmt_job_running'] = pd.to_datetime(df['gmt_job_running'], format='%Y/%m/%d %H:%M')
df.loc[:, 'gmt_job_finished'] = pd.to_datetime(df['gmt_job_finished'], format='%Y/%m/%d %H:%M')
```

```python
df.loc[:, 'duration'] = df['gmt_job_finished'] - df['gmt_job_running']
df['duration'] = pd.to_timedelta(df['duration'], unit='s')
plt.ecdf(df['duration'])
df['duration']
```

```python
from collections import OrderedDict

def count_jobs_at_each_moment(jobs: pd.DataFrame):
    # Creating a list to store start and end events
    events = []
    for _, row in jobs.iterrows():
        start = row['gmt_job_running']
        end = row['gmt_job_finished']
        events.append((start, 'start'))
        events.append((end, 'end'))

    # Sort events, first by time, then by type ('end' before 'start' at the same time)
    events.sort(key=lambda x: (x[0], x[1] == 'start'))

    # Counting the number of jobs at each moment
    current_active_jobs = 0
    results = {}
    for time, type in events:
        if type == 'start':
            current_active_jobs += 1
        else:
            current_active_jobs -= 1
        # We store the job count after processing the event to avoid counting ends prematurely
        if time in results:
            results[time] = max(results[time], current_active_jobs)
        else:
            results[time] = current_active_jobs
            
    sorted_job_counts = OrderedDict(sorted(results.items()))

    return sorted_job_counts

running_jobs = count_jobs_at_each_moment(df)
running_jobs
```

```python
plt.step(running_jobs.keys(), running_jobs.values(), where='post')
```
